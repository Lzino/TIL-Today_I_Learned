#### Binomial distribution

```
# https://towardsdatascience.com/fun-with-the-binomial-distribution-96a5ecabf65b

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Random variable
trials = 1000

# Number of independent experiments in each trial
number = 100

# Probability of success for each experiment
p = 0.5

def run_binom(trials, n, p) :
  heads = []
  for i in range (trials) :
    tosses = [np.random.random() for i in range(n)]
    heads.append(len([i for i in tosses if i>=0.50]))
  return heads
```

```
heads = run_binom(trials, number, p)
```

```
# Visualization
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.distplot(heads, bins = 11, label = 'Binomial result')
ax.set_xlabel("Number of Heads",fontsize=16)
ax.set_ylabel("Frequency",fontsize=16)
```

#### Categorical distribution

```
# https://datascienceschool.net/view-notebook/354eb1d0e307423697ed805df07b96a5/

import scipy as sp
import pandas as pd

mu = [0.1, 0.1, 0.1, 0.1, 0.3, 0.3]
rv = sp.stats.multinomial(1, mu) # for categorical
```

```
xx = np.arange(1, 7)
xx_ohe = pd.get_dummies(xx) # dummy variable. That means one hot encoding

plt.bar(xx, rv.pmf(xx_ohe.values))
plt.ylabel("P(x)")
plt.xlabel("The value of sample")
plt.title("Pmf of categorical distribution")
```

```
# You can also use this as a substitute.
np.random.seed(1)
X = rv.rvs(200)
y = X.sum(axis = 0) / float(len(X))

plt.bar(np.arange(1,7), y)
plt.ylabel("P(x)")
plt.xlabel("The value of sample")
plt.title("Simulation of categorical distribution")
plt.show()
```

#### Gaussian distribution

```
# https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.normal.html

# mean & standard
mu, sigma = 0, 0.1

# random sampling
s = np.random.normal(mu, sigma, 1000)

# samples's histogram
count, bins, ignored = plt.hist(s, 30, density =True)

def gaussian(mu, sigma, bins):
  result = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( -(bins - mu)**2 / (2 * sigma**2))
  return result

plt.plot(bins, gaussian(mu, sigma, bins), linewidth = 2, color ='r')
plt.show()
```
